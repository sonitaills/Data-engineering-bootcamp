## Criar uma função método que faz todo o tratamento do tweet e converte para um DataFrame Pandas!

```python
# Nem sempre a estrutura de dados vem correta, por isso foi implementado o try / except.
# A maneira como o professor implementou, porém não é muito bacana... ela ignoraria o tweet por completo caso desse erro
# Para corrigir isso não é muito simple, o tweet 30 por exemplo, retorna erros pq:
#--> coluna retweeted_status não existe, erro no drop columns
#--> tweet['entities']['user_mentions'] tmb não existe.
#Muitas outras coisas podem conter erros. Uma maneira de corrigir o drop columns seria checar quais deses nomes
#não estão em df_tratado.columns e tirar os que aparecem.
#para o entities, poderia ter um try except somente para o for loop.

def tweet_para_df(tweet):
    try:
        df_tratado = pd.DataFrame(tweet).reset_index(drop=True).iloc[:1]
        df_tratado.drop(columns = ['quote_count','reply_count', 'retweet_count', 'favorite_count','favorited', 
                                   'retweeted','user','entities','retweeted_status'], inplace = True)
        df_tratado['user_id'] = tweet['user']['id']
        df_tratado['user_id_str'] = tweet['user']['id_str']
        df_tratado['user_screen_name'] = tweet['user']['screen_name']
        df_tratado['user_location'] = tweet['user']['location']
        df_tratado['user_description'] = tweet['user']['description']
        df_tratado['user_protected'] = tweet['user']['protected']
        df_tratado['user_verified'] = tweet['user']['verified']
        df_tratado['user_followers_count'] = tweet['user']['followers_count']
        df_tratado['user_friends_count'] = tweet['user']['friends_count']
        df_tratado['user_created_at'] = tweet['user']['created_at']

        user_mentions =[]

        for i in range(len(tweet['entities']['user_mentions'])):
            dicionariobase = tweet['entities']['user_mentions'][i].copy() #copy utilizado para não alterar o dado original
            dicionariobase.pop('indices',None) # não queremos a chave indive. Se ela não existir, o None vai evitar erro no programa
            df = pd.DataFrame(dicionariobase,index =[0])
            df = df.rename(columns= {
                'screen_name':'entities_screen_name',
                'name':'entities_name',
                'id':'entities_id',
                'id_str':'entities_id_str'
            })
            user_mentions.append(df)
###################################Se carregasse de 1 por 1 no dataframe, o código iria ficar muito mais lento!
        dfs =[]
        for i in user_mentions:
            dfs.append(
            pd.concat([df_tratado.copy(),i],axis=1)
            )
        df_final = pd.concat(dfs,ignore_index=True)
    except:
        return None
    return df_final
```

## Iterando e transformando todos os tweets

```python
%%time
#Aqui a mesma ideia... cria uma lista e depois carrega ela de uma vez em um df
parseados = [tweet_para_df(tweet) for tweet in parsed_tweets]
```

```python
print(f'Quantidade de dados parseados = {len(parseados)}')
#Eliminar as posições vazias da lista (a 30, por exemplo, é vazia!)
clean_parseados = [i for i in parseados if i is not None]
print(f'Quantidade de dados parseados não nulos = {len(clean_parseados)}')
print(f'Dados descartados = {-len(clean_parseados)+len(parseados)}')
```

```python
tratado = pd.concat(clean_parseados,ignore_index=True)
```

```python
g = tratado.groupby('text').agg({'text':'count'})
res = g.apply(lambda x: x.sort_values(ascending=False))
```
